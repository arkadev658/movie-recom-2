```python
import numpy as np
import pandas as pd
```

```python
credits=pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv')
credits.head()
```

```python
movies_df=pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
movies_df.head()
```

```python
credits.rename(columns={'movie_id':'id'},inplace=True)
credits.head()
```

```python
new_df=movies_df.merge(credits,on='id')
new_df.head()
```

```python
new_df.info()
```

```python
# dropping unnnecessary columns
# important columns are: generes,id, keywords, original title, overview, popularity, cast, crew
new_df=new_df[['genres','id', 'keywords', 'original_title', 'overview', 'popularity', 'cast', 'crew']]
```

```python
#data cleaning

#is null value there ?
new_df.isna().sum()
new_df.dropna(inplace=True)

#is there any duplicate value ?
new_df.duplicated().sum()
```

```python
# formatting the genre
import ast
def convgenre(obj):
    L=[]
    for elem in ast.literal_eval(obj):
        L.append(elem['name'])
    return L
new_df['genres']=new_df['genres'].apply(convgenre)
```

```python
# formatting tehe keywords
def convkey(obj):
    L=[]
    for elem in ast.literal_eval(obj):
        L.append(elem['name'])
    return L
new_df['keywords']=new_df['keywords'].apply(convgenre)   
```

```python
# overview
new_df['overview']=new_df['overview'].apply(lambda x: x.split(' '))
```

```python
# cast (only first 3 cast has been taken)
def convcast(obj):
    L=[]
    count=0
    for elem in ast.literal_eval(obj):
        if count<3:
            L.append(elem['character'])
            count+=1
        else:
            break
    return L
new_df['cast']=new_df['cast'].apply(convcast)
```

```python
#crew (only the name of the director can be taken)
def convcrew(obj):
    L=[]
    count=0
    for elem in ast.literal_eval(obj):
        if elem['job']=='Director':
            L.append(elem['name'])
            break
    return L
new_df['crew']=new_df['crew'].apply(convcrew)
```

```python
# removing spaces for confusion
for column in ['genres','keywords','overview','crew','cast']:
    new_df[column]=new_df[column].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
# create new column tag
new_df['tags']=new_df['genres']+new_df['keywords']+new_df['overview']+new_df['cast']+new_df['crew']
new_df['tags']=new_df['tags'].apply(lambda x: ' '.join(x))
```

```python
new_df
```

```python
#stemming
from nltk.stem import PorterStemmer
ste=PorterStemmer()
def stemming(obj):
    y=[]
    for word in obj.split(' '):
        y.append(ste.stem(word))
    return ' '.join(y)
new_df['tags']=new_df['tags'].apply(stemming)
```

```python
#vectorization of tags
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer(min_df=3,max_features=5000, strip_accents='unicode', analyzer='word',
                   ngram_range=(1,3),stop_words='english')
```

```python
vectors=vec.fit_transform(new_df['tags']).toarray()
```

```python
# similarity finding
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
#recommendation
def recommend(movie):
    movie_index=new_df[new_df['original_title']==movie].index[0]
    sim=list(enumerate(similarity[movie_index]))
    recom=sorted(sim,key=(lambda x: x[1]),reverse=True)[1:6]
    for i in recom:
        print(new_df.iloc[i[0]]['original_title'])
```

```python
#test
recommend('Terminator 2: Judgment Day')
```
